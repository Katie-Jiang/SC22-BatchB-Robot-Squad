## Breast Cancer



```python
breast_cancer = pd.read_csv("Data/breast_cancer.csv")
breast_cancer.head(5)
```

```python
breast_cancer.shape
```

```python
plt.hist(breast_cancer['radius_mean'], bins=25, alpha=0.45, color='red')
plt.hist(breast_cancer['diagnosis'], bins=25, alpha=0.45, color='blue')
plt.xlabel('radius_mean')
plt.ylabel('number of people')
```

```python
plt.hist(breast_cancer[breast_cancer['diagnosis'] == "M"].head()['radius_mean'], bins=25, alpha=0.45, color='red')
plt.hist(breast_cancer[breast_cancer['diagnosis'] == "B"].head()['radius_mean'], bins=25, alpha=0.45, color='blue')
plt.xlabel('radius_mean')
plt.ylabel('number of people')
```

```python
plt.hist(breast_cancer[breast_cancer['diagnosis'] == "M"].head()['perimeter_mean'], bins=25, alpha=0.45, color='red')
plt.hist(breast_cancer[breast_cancer['diagnosis'] == "B"].head()['perimeter_mean'], bins=25, alpha=0.45, color='blue')
plt.xlabel('perimter_mean')
plt.ylabel('number of people')
```

```python
breast_cancer.plot.scatter(x= 'area_mean',y='area_worst')
```

```python
breast_cancer.plot.scatter(x= 'radius_mean',y='perimeter_mean')
```

```python
breast_cancer.plot.scatter(x= 'smoothness_mean',y='concavity_mean', title="Concavity Mean vs. smoomthness mean")
```

```python
breast_cancer.plot.scatter(x= 'area_mean',y='diagnosis')
```

```python
fig = px.scatter(breast_cancer, x="smoothness_mean", y="diagnosis", title="Smoothness Mean vs. Diagnosis", color="smoothness_mean")
fig.show()
```

```python
fig = px.scatter(breast_cancer, x="area_mean", y="diagnosis", color="area_mean", title="Area Mean vs. Diagnosis")
fig.show()
```
